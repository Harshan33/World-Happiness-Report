# Data Visulations

## Box Plot of Ladder Score






 

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y='Ladder score', data=df2024)
plt.title('Box Plot of Ladder score')
plt.ylabel('Ladder Score')
plt.show()
```

## Ladder Score Distribution with Histogram

```python
plt.figure(figsize =(10,6))
sns.histplot(df2024['Ladder score'],bins = 20 , kde=True)
plt.title('Histogram of Ladder score')
plt.xlabel('Ladder score')
plt.ylabel('Frequency')
plt.show()
```
